## Simulation results

This file contains single Au-197 ion shootting towards a MVTX ladder. The ion kinematics is 

```c++

    INPUTGENERATOR::Gun[0]->AddParticle(1000791970, 0, 0, 19700);
    INPUTGENERATOR::Gun[0]->set_vtx(-3.264547, 0.5698622, -30);

```

````python
file = ROOT.TFile.Open("data/G4sPHENIX_DSTReader_AuOnMVTX_FTFP_BERT_HP.root")
T = file.Get("T")
# T.Show(0)
````

# 3D display, Event 1

````python
   

c1 = ROOT.TCanvas( '3DHits', '3DHits', 900, 900)

T.Draw("G4HIT_MVTX.get_avg_y():G4HIT_MVTX.get_avg_z():G4HIT_MVTX.get_avg_x()","Entry$==1")
ROOT.gPad.GetPrimitive("htemp").GetXaxis().SetTitle("G4Hit x [cm]")
ROOT.gPad.GetPrimitive("htemp").GetYaxis().SetTitle("G4Hit z [cm]")
ROOT.gPad.GetPrimitive("htemp").GetZaxis().SetTitle("G4Hit y [cm]")

ROOT.TPolyMarker3D() 

c1.Draw()
````
